```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pylab as plt
import math

from sklearn.linear_model import LogisticRegression

%matplotlib inline
pd.options.display.max_columns = None
```

```python
test_raw = pd.read_csv("test.csv")
train_raw = pd.read_csv("train.csv")
```

```python
train_raw.head()
```

```python
def clean_data(data):
    return data

def expand_data(data):
    data['age'] = data['YrSold'] - data['YearBuilt'] 
    data['age2'] = data['age'] ** 2
    
    gen_dummies = pd.get_dummies(data['MSZoning'])
    gen_dummies.rename(columns={"C (all)": "C"})
    
    data = pd.concat([data, gen_dummies], 
           axis=1)
    return data
    
train = expand_data(clean_data(train_raw))
test  = expand_data(clean_data(test_raw))
```

```python

gen_dummies = pd.get_dummies(train['MSZoning'])
gen_dummies.columns
```

```python
formula = "SalePrice ~ age + age2 + LotArea + GarageArea + FV + RH + RL + RM - 1"
lm = smf.ols(formula = formula, data = train).fit()
lm.summary()
```
